**1. Latar Belakang**

Saham NVDA (NVIDIA Corporation) merupakan salah satu saham teknologi yang paling populer dan volatil di pasar saham. Mengingat pergerakan harga saham yang signifikan dalam beberapa tahun terakhir, banyak investor dan trader yang tertarik untuk mengetahui prediksi harga saham ini di masa mendatang.

Forecasting harga saham NVDA dilakukan dengan tujuan untuk:

- Mengidentifikasi tren harga saham NVDA untuk mendukung keputusan investasi dan trading serta mengurangi risiko kerugian.

Model Prophet digunakan dalam prediksi ini karena kemampuannya untuk menangkap pola musiman dan tren yang kompleks, serta sensitivitas terhadap perubahan tren yang dapat disesuaikan dengan parameter tertentu. Meskipun terdapat beberapa kelemahan dalam prediksi fluktuasi jangka pendek, model ini memberikan pandangan yang kuat mengenai tren jangka panjang harga saham NVDA.

**2. Analytical Approach**

Pendekatan yang digunakan dalam forecasting ini adalah model **Prophet**, yang dirancang untuk memprediksi data time series dengan pola musiman dan perubahan tren yang kompleks. Langkah-langkah utama yang dilakukan dalam modeling ini adalah sebagai berikut:

1. **Preprocessing Data:** Data harga penutupan saham NVDA dari tahun 2024 digunakan, dengan transformasi tanggal dan perubahan nama kolom untuk keperluan modeling.
2. **Pembagian Data:** Data dibagi menjadi 80% data training dan 20% data testing untuk menguji kemampuan model dalam memprediksi harga saham.
3. **Tuning Model:** Model Prophet dituning menggunakan parameter `changepoint_prior_scale`, `seasonality_prior_scale`, dan `holidays_prior_scale` untuk mengatur sensitivitas model terhadap perubahan tren dan pola musiman.
4. **Evaluasi Model:** Model dievaluasi menggunakan metrik MAE dan RMSE untuk mengetahui seberapa akurat prediksi dibandingkan dengan nilai aktual.

**3. Hasil dan Analisis**

Dari hasil prediksi, model Prophet mampu menangkap tren besar dalam pergerakan harga saham NVDA. Namun, terdapat beberapa deviasi pada periode dengan volatilitas tinggi. Meskipun begitu, model ini memberikan gambaran umum yang akurat mengenai arah pasar, yang sangat berguna untuk strategi investasi dan trading jangka panjang.

Evaluation Metric
- **Mean Absolute Error (MAE):** Menunjukkan seberapa jauh prediksi model dari nilai aktual secara rata-rata.

Visualisasi
Berikut adalah beberapa visualisasi yang menunjukkan performa model dalam memprediksi harga saham NVDA:

   - Rata-rata bergerak 30-hari menunjukkan pergerakan harga saham yang lebih mulus dan lebih cepat merespons perubahan harga dibandingkan dengan rata-rata bergerak 90-hari. Pada beberapa titik, rata-rata ini lebih dekat mengikuti fluktuasi harga aktual.

## **Changepoints in NVDA Stock Price**

```python
import ruptures as rpt

signal = df['Close'].values

algo = rpt.Pelt(model="rbf").fit(signal)

result = algo.predict(pen=5)  
valid_cp = [cp for cp in result if cp < len(df)]

plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Close'], label='Close Price')

for cp in valid_cp:
    plt.axvline(x=df['Date'].iloc[cp], color='red', linestyle='--', label='Changepoint')

plt.title('Changepoints in NVDA Stock Price')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()

print("Changepoints yang valid (Index dan Date):")
for cp in valid_cp:
    print(f'Index: {cp}, Date: {df["Date"].iloc[cp]}')
```

1. **Mengurangi Dampak Anomali:**
   - Dengan memotong data dari sebelum tahun 2024 kita dapat menghindari pengaruh data yang terlalu volatil atau anomali dari tahun-tahun sebelumnya yang mungkin tidak relevan lagi untuk periode prediksi pendek.
   
2. **Fokus pada Tren Terkini:**
   - Pemotongan data memungkinkan model untuk lebih fokus pada pola dan tren terbaru, yang biasanya lebih relevan untuk prediksi jangka pendek. Hal ini penting karena pola harga yang lebih lama mungkin tidak lagi mencerminkan kondisi pasar saat ini.

3. **Mengurangi Overfitting:**
   - Data historis yang terlalu panjang dengan banyak perubahan tren bisa membuat model overfitting, yaitu terlalu mengikuti noise dari data historis. Dengan menggunakan data terbaru, model bisa lebih general dan memberikan hasil prediksi yang lebih baik.

4. **Mengenali Pola Musiman Terbaru:**
   - Tahun 2024 mungkin memiliki pola musiman atau faktor spesifik yang berbeda dari tahun-tahun sebelumnya. Dengan fokus pada data terbaru, Anda bisa lebih mudah mengenali pola-pola ini dan memasukkannya ke dalam model prediksi.

5. **Menggunakan Changepoints untuk Pembagian Data:**
   - Berdasarkan changepoints yang terlihat, perubahan tren di awal 2024 bisa menjadi dasar untuk menentukan titik pemotongan data. Ini memastikan bahwa model memiliki informasi yang cukup dari tren yang baru tetapi tidak terlalu dipengaruhi oleh perubahan yang lebih lama.

# Default Tunning Model

```python
from prophet import Prophet
from sklearn.metrics import mean_squared_error, mean_absolute_error

df2 = pd.read_csv('NVDA_2024.csv') 
df2['Date'] = pd.to_datetime(df2['Date'])
df2 = df2.rename(columns={'Date': 'ds', 'Close': 'y'})

train_size = int(0.8 * len(df2))
train_df2 = df2[:train_size]
test_df2 = df2[train_size:]


model = Prophet(
   
    yearly_seasonality=True,
    weekly_seasonality=True,
    daily_seasonality=False,
)

model.add_seasonality(name='monthly', period=30.5, fourier_order=5)

model.fit(train_df2)

future = model.make_future_dataframe(periods=3 * 30, freq='D') 
forecast = model.predict(future)

test_forecast = forecast[forecast['ds'].isin(test_df2['ds'])]

test_merged = test_df2.merge(test_forecast[['ds', 'yhat']], on='ds', how='left')

residuals = test_merged['y'] - test_merged['yhat']

mae = mean_absolute_error(test_merged['y'], test_merged['yhat'])

start_forecast_after_test = test_df2['ds'].max() + pd.Timedelta(days=1)


plt.figure(figsize=(12, 6))
plt.plot(df2['ds'], df2['y'], label='Actual', color='black')
plt.plot(forecast['ds'], forecast['yhat'], label='Forecast', color='blue')
plt.fill_between(forecast['ds'], forecast['yhat_lower'], forecast['yhat_upper'], color='blue', alpha=0.2)
plt.axvline(x=df2.iloc[train_size]['ds'], color='r', linestyle='--', label='Train-Test Split')
plt.axvline(x=start_forecast_after_test, color='green', linestyle='--', label='Start Forecast After Test')
plt.title(f'Prophet Forecast vs Actual (Default Parameters)')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.show()


print(f'Mean Absolute Error (MAE): {mae}')
print('-' * 50)
```

## 1. **Pemilihan Parameter Model:**
   - Parameter `changepoint_prior_scale` (CPS), `seasonality_prior_scale` (SPS), dan `holidays_prior_scale` (HPS) digunakan untuk mengatur sensitivitas model terhadap perubahan tren, pola musiman, dan liburan. Pada tuning kali ini:
     - **CPS = 0.06:** Model lebih fleksibel dalam menangkap perubahan tren tanpa terlalu overfitting.
     - **SPS = 0.15:** Sensitivitas terhadap pola musiman cukup tinggi, yang cocok untuk menangkap pola musiman saham seperti yang teridentifikasi sebelumnya.
     - **HPS = 1:** Tidak terlalu banyak pengaruh karena liburan tidak signifikan dalam data ini.

## 2. **Hasil Evaluasi Model:**
   - **MAE (Mean Absolute Error):** MAE menunjukkan seberapa jauh rata-rata prediksi menyimpang dari nilai aktualnya. Nilai yang lebih rendah berarti model cukup baik dalam memprediksi data uji.
   - **RMSE (Root Mean Squared Error):** RMSE memberikan informasi tentang seberapa besar error model dalam skala harga saham. Nilai ini lebih sensitif terhadap error besar, sehingga lebih berguna dalam mengidentifikasi deviasi ekstrem.

## 3. **Visualisasi Hasil Prediksi vs. Actual:**
   - **Periode Prediksi:** Pada grafik hasil prediksi, terlihat bahwa model cukup baik dalam mengikuti tren data aktual selama periode pengujian. Area bayangan biru menunjukkan interval kepercayaan, dan tampaknya sebagian besar data aktual berada dalam rentang ini, yang menunjukkan tingkat kepercayaan model yang baik.
   - **Forecast Setelah Data Test:** Seperti yang dibahas pada insight sebelumnya, pemotongan data dari awal 2024 membantu model lebih fokus pada tren terbaru. Ini terlihat pada visualisasi, di mana model bisa menangkap pola kenaikan atau penurunan harga lebih akurat di periode terbaru.

## 4. **Residuals:**
   - Visualisasi residual menunjukkan bahwa error model cukup terdistribusi merata tanpa pola yang jelas, yang menunjukkan bahwa model telah menangkap sebagian besar pola data yang ada tanpa bias sistematis.

Dengan parameter tuning ini, model Prophet telah disesuaikan untuk menangkap pola musiman dan tren yang lebih sensitif terhadap perubahan kecil, seperti yang diidentifikasi pada analisis rolling mean dan changepoints sebelumnya.

```python
train_size_tunned = int(0.8 * len(df2))
train_df2_tunned = df2[:train_size_tunned]
test_df2_tunned = df2[train_size_tunned:]

changepoint_prior_scale = 0.06
seasonality_prior_scale = 0.15
holidays_prior_scale = 1

model = Prophet(
    changepoint_prior_scale=changepoint_prior_scale,  
    seasonality_prior_scale=seasonality_prior_scale,  
    holidays_prior_scale=holidays_prior_scale,       
    yearly_seasonality=True,
    weekly_seasonality=True,
    daily_seasonality=False,
)

model.add_seasonality(name='monthly', period=30.5, fourier_order=5)

model.fit(train_df2_tunned)

future = model.make_future_dataframe(periods=4 * 30, freq='D') 
forecast = model.predict(future)

test_forecast = forecast[forecast['ds'].isin(test_df2_tunned['ds'])]

test_merged = test_df2_tunned.merge(test_forecast[['ds', 'yhat']], on='ds', how='left')

residuals = test_merged['y'] - test_merged['yhat']

mae = mean_absolute_error(test_merged['y'], test_merged['yhat'])

start_forecast_after_test = test_df2_tunned['ds'].max() + pd.Timedelta(days=1)


plt.figure(figsize=(12, 6))
plt.plot(df2['ds'], df2['y'], label='Actual', color='black')
plt.plot(forecast['ds'], forecast['yhat'], label='Forecast', color='blue')
plt.fill_between(forecast['ds'], forecast['yhat_lower'], forecast['yhat_upper'], color='blue', alpha=0.2)
plt.axvline(x=df2.iloc[train_size_tunned]['ds'], color='r', linestyle='--', label='Train-Test Split')
plt.axvline(x=start_forecast_after_test, color='green', linestyle='--', label='Start Forecast After Test')
plt.title(f'Prophet Forecast vs Actual (CPS={changepoint_prior_scale}, SPS={seasonality_prior_scale}, HPS={holidays_prior_scale})')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(test_merged['ds'], residuals, label='Residuals', color='red')
plt.title(f'Residuals of the Model (CPS={changepoint_prior_scale}, SPS={seasonality_prior_scale}, HPS={holidays_prior_scale})')
plt.xlabel('Date')
plt.ylabel('Residual')
plt.legend()
plt.show()

print(f'Parameters: CPS={changepoint_prior_scale},  SPS={seasonality_prior_scale}, HPS={holidays_prior_scale}')
print(f'Mean Absolute Error (MAE): {mae}')
print('-' * 50)
```
